```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Re-import the dataset due to the reset
df = pd.read_excel('final_dataset.xlsx')

# Prepare the data
df = df.drop('Unnamed: 0', axis=1)
df = df.set_index('Index')
Y = df['Sea_level']
X = df.drop('Sea_level', axis=1)

# Add a constant to the independent variable set (for the intercept)
X = add_constant(X)

# Calculate VIF for each explanatory variable
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# VIF calculation can be infinity if a feature is a perfect linear combination of others.
# To avoid division by zero in VIF calculation, we add a small value to the denominator.
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif_data
```

```python
import statsmodels.api as sm
import numpy as np

# Step 1: Initial OLS Regression
model = sm.OLS(Y, X_final)
results = model.fit()
residuals = results.resid

# ols_results_final

# Initialize variables for the iterative process
rho = 1  # Initial guess for the autocorrelation coefficient
tolerance = 1e-6  # Convergence tolerance
max_iter = 100  # Maximum number of iterations
iter_count = 0

while iter_count < max_iter:
    iter_count += 1
    
    # Step 2: Estimate the first-order autocorrelation (rho) of the residuals
    rho_new = np.corrcoef(residuals[:-1], residuals[1:])[0, 1]

    # Check for convergence
    if abs(rho_new - rho) < tolerance:
        break
    
    rho = rho_new

    # Step 3: Transform the variables
    Y_transformed = Y[1:] - rho * Y[:-1]
    X_transformed = X_final[1:] - rho * X_final[:-1]

    # ... previous steps to estimate rho and transform variables ...

    # Drop NaNs from the transformed variables
    Y_transformed = Y_transformed.dropna()
    X_transformed = X_transformed.dropna()

    # Check for and handle infinite values if they exist
    if np.isinf(X_transformed).values.any() or np.isinf(Y_transformed).values.any():
        X_transformed.replace([np.inf, -np.inf], np.nan, inplace=True)
        Y_transformed.replace([np.inf, -np.inf], np.nan, inplace=True)
        X_transformed.dropna(inplace=True)
        Y_transformed.dropna(inplace=True)

    # Step 4: Rerun OLS regression on the transformed variables
    model_transformed = sm.OLS(Y_transformed, X_transformed)
    results_transformed = model_transformed.fit()
    residuals = results_transformed.resid
    

# Print the final results
print("Converged after", iter_count, "iterations")
print("rho: ", rho)
print(results_transformed.summary())
print("Durbin-Watson statistic:", durbin_watson(residuals))
```

```python
# Drop the first observation since it will be NaN after the transformation
Y_dropped = Y.drop(index=Y.index[0])
X_dropped = X_final.drop(index=X_final.index[0])

# Perform the Prais-Winsten transformation on all variables
Y_transformed = Y_dropped - rho * Y.shift(1).drop(index=Y.index[0])
X_transformed = X_dropped.subtract(rho * X_final.shift(1).drop(index=X_final.index[0]))

# Perform OLS regression on the transformed data
prais_winsten_model = sm.OLS(Y_transformed, X_transformed)
prais_winsten_results = prais_winsten_model.fit()

# Get a summary of the OLS model with Prais-Winsten transformed data
prais_winsten_results_summary = prais_winsten_results.summary()
prais_winsten_results_summary
```

```python
import numpy as np
from statsmodels.tsa.stattools import acf
from statsmodels.regression.linear_model import GLS
# Estimate the first-order autocorrelation coefficient
acf_coefficient = acf(ols_results_final.resid, fft=False, nlags=1)[1]

# Number of observations
n = len(ols_results_final.resid)

# Constructing a Toeplitz matrix for the GLS omega matrix
toeplitz_matrix = acf_coefficient ** np.abs(np.subtract.outer(np.arange(n), np.arange(n)))

# Apply the GLS model using the Toeplitz matrix as the omega matrix
gls_toeplitz_model = GLS(Y, X_final, sigma=toeplitz_matrix)
gls_toeplitz_results = gls_toeplitz_model.fit()

# Get a summary of the GLS model results with Toeplitz omega matrix
gls_toeplitz_results_summary = gls_toeplitz_results.summary()
gls_toeplitz_results_summary
```
